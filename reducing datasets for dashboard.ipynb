```python
import pandas as pd
```

```python
kmeansDf = pd.read_csv('.\dataset_kmeans_clusters.csv')
```

```python
kmeansDf.columns
```

```python
del kmeansDf['book_id']
del kmeansDf['author_id']
del kmeansDf['Unnamed: 0']
```

```python
kmeansDf
```

```python
kmeansDf = kmeansDf.drop_duplicates(subset=['book_title'])
```

```python
kmeansDf
```

```python
kmeansDf['author_average_rating'].unique()
```

```python
kmeansDf = kmeansDf[kmeansDf['book_average_rating'] >= 0.6]
```

```python
kmeansDf = kmeansDf[kmeansDf['author_average_rating'] >= 0.6]
```

```python
kmeansDf
```

```python
kmeansDf.to_csv('dataset_kmeans_clusters.csv')
```
